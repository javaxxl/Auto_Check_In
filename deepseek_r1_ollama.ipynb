```python
#@title 下载 DeepSeek 模型

model_name = "codellama:7b" # 替换成 DeepSeek-R1 的实际模型名称

!ollama pull {model_name}

print(f"codellama 模型 {model_name} 下载完成！")
```

```python
#@title 安装 Gradio
!pip install -q gradio requests

import gradio as gr
import requests

# Ollama API 端点 (默认为本地)
OLLAMA_API_URL = "http://localhost:11434/api/generate"

# 创建 Gradio 界面
def generate_text(prompt):
  data = {
      "prompt": prompt,
      "model": model_name,  # 使用下载的模型名称
      "stream": False  # Colab 上 stream 可能会有问题，先关闭
  }
  response = requests.post(OLLAMA_API_URL, json=data, stream=False) # Ensure stream=False for Colab
  if response.status_code == 200:
    return response.json()['response']
  else:
    return f"Error: {response.status_code} - {response.text}"

# 定义 Gradio 界面
iface = gr.Interface(
    fn=generate_text,
    inputs=gr.Textbox(lines=7, placeholder="输入你的提示词..."),
    outputs=gr.Textbox(),
    title="codellama on Ollama",
    description="在 Google Colab 上使用 Ollama Server 运行 codellama 模型。",
)

# 启动 Gradio 界面
iface.launch(share=True)
```
